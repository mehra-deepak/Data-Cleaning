```python
import nltk

File = open("new_file.csv", encoding="utf8") 
lines = File.read() 
sentences = nltk.sent_tokenize(lines) 
nouns = [] 

for sentence in sentences:
     for word,pos in nltk.pos_tag(nltk.word_tokenize(str(sentence))):
            if ( pos == 'NNP'):
                nouns.append(word)
```
